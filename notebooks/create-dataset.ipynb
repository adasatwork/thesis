```python
from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
hqs_train1 = pd.read_excel("https://github.com/abachaa/MeQSum/raw/refs/heads/master/MeQSum_ACL2019_BenAbacha_Demner-Fushman.xlsx")
hqs_val = pd.read_excel("https://github.com/abachaa/MEDIQA2021/raw/refs/heads/main/Task1/MEDIQA2021-Task1-QuestionSummarization-ValidationSet.xlsx")
hqs_test = pd.read_excel("https://github.com/abachaa/MEDIQA2021/raw/refs/heads/main/Task1/MEDIQA2021-Task1-TestSet-ReferenceSummaries.xlsx")
```

```python
hqs_train1.drop("File", axis=1, inplace=True)
```

```python
hcmsum1 = pd.read_json("https://github.com/zekaouinoureddine/MedQSum/raw/refs/heads/master/data/hcm_sum/train.json")
hcmsum2 = pd.read_json("https://github.com/zekaouinoureddine/MedQSum/raw/refs/heads/master/data/hcm_sum/valid.json")

chqsum1 = pd.read_json("https://github.com/zekaouinoureddine/MedQSum/raw/refs/heads/master/data/chq_sum/train.json")
chqsum2 = pd.read_json("https://github.com/zekaouinoureddine/MedQSum/raw/refs/heads/master/data/chq_sum/valid.json")
```

```python
hqs_train2 = pd.concat([hcmsum1, hcmsum2], ignore_index=True)
hqs_train3 = pd.concat([chqsum1, chqsum2], ignore_index=True)
```

```python
hqs_train = pd.concat([hqs_train1, hqs_train2, hqs_train3], ignore_index=True)
```

```python
idx_to_remove = hqs_train[(hqs_train.apply(lambda x: len(x["CHQ"].split()), axis=1) > 400)].index
```

```python
hqs_train.drop(idx_to_remove, inplace=True)
```

```python
hqs_train.info()
```

```python
hqs_train.rename(columns={"CHQ": "source", "Summary": "summary"}, inplace=True)
```

```python
hqs_val.drop(["q_id", "Question Focus", "Question Type"], axis=1, inplace=True)
hqs_val.rename(columns={"NLM Question": "source", "Summary": "summary"}, inplace=True)
```

```python
hqs_test.drop("q_id", axis=1, inplace=True)
hqs_test.rename(columns={"NLM Question": "source", "Summary": "summary"}, inplace=True)
```

```python
hqs_train_sft, hqs_train_xpo = train_test_split(hqs_train, test_size=0.4, random_state=42)
```

```python
hqs_data_path = Path.cwd().parent / "data" / "hqs"

hqs_train.to_json(hqs_data_path / "train.jsonl", orient="records", lines=True)
hqs_val.to_json(hqs_data_path / "dev.jsonl", orient="records", lines=True)
hqs_test.to_json(hqs_data_path / "test.jsonl", orient="records", lines=True)

hqs_train_sft.to_json(hqs_data_path / "train-sft.jsonl", orient="records", lines=True)
hqs_train_xpo.to_json(hqs_data_path / "train-xpo.jsonl", orient="records", lines=True)
```
