```python
from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
hqs_train = pd.read_excel("https://github.com/abachaa/MeQSum/raw/refs/heads/master/MeQSum_ACL2019_BenAbacha_Demner-Fushman.xlsx")
hqs_val = pd.read_excel("https://github.com/abachaa/MEDIQA2021/raw/refs/heads/main/Task1/MEDIQA2021-Task1-QuestionSummarization-ValidationSet.xlsx")
hqs_test = pd.read_excel("https://github.com/abachaa/MEDIQA2021/raw/refs/heads/main/Task1/MEDIQA2021-Task1-TestSet-ReferenceSummaries.xlsx")
```

```python
hqs_train.drop("File", axis=1, inplace=True)
hqs_train.rename(columns={"CHQ": "source", "Summary": "ref"}, inplace=True)
```

```python
hqs_val.drop(["q_id", "Question Focus", "Question Type"], axis=1, inplace=True)
hqs_val.rename(columns={"NLM Question": "source", "Summary": "ref"}, inplace=True)
```

```python
hqs_test.drop("q_id", axis=1, inplace=True)
hqs_test.rename(columns={"NLM Question": "source", "Summary": "ref"}, inplace=True)
```

```python
hqs_train_p1, hqs_train_p2 = train_test_split(hqs_train, test_size=0.2, random_state=42)
```

```python
hqs_data_path = Path.cwd().parent / "data" / "hqs"

hqs_train.to_json(hqs_data_path / "hqs-train.jsonl", orient="records", lines=True)
hqs_val.to_json(hqs_data_path / "hqs-val.jsonl", orient="records", lines=True)
hqs_test.to_json(hqs_data_path / "hqs-test.jsonl", orient="records", lines=True)

hqs_train_p1.to_json(hqs_data_path / "hqs-train-p1.jsonl", orient="records", lines=True)
hqs_train_p2.to_json(hqs_data_path / "hqs-train-p2.jsonl", orient="records", lines=True)
```
